```python
#numeric variables
numeric_variables=train.select_dtypes(include=[np.number])
numeric_variables.columns
```

```python
numeric_variables.head()
```

```python
#year variables
year_variables=[feature  for feature in numeric_variables if 'Yr' in feature or 'Year' in feature]
#year_variables=[]
#for feature in numeric_variables:
#    if 'Yr' in feature or 'Year' in feature:
#        year_variables.append(feature)
year_variables
```

```python
for feature in year_variables:
    print(feature,train[feature].unique())
```

```python
for feature in year_variables:
    if feature !='YrSold':
        data= train.copy()
        data[feature]=data["YrSold"]-data[feature]
        
        
        #comparing Al year and sales prices
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Discrete varuables
discrete_variables=[feature for feature in numeric_variables if len(train[feature].unique())<25 and feature not in year_variables +['ID']]
print("Discrete Variables Count: {}".format(len(discrete_variables)))
discrete_variables
```

```python
train[discrete_variables].head()
```

```python
for feature in discrete_variables:
    data=train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#continous variables
continous_variables=[feature for feature in numeric_variables if feature not in discrete_variables+year_variables+['Id']]
continous_variables,len(continous_variables)
```

```python
for feature in continous_variables:
    data=train.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```
